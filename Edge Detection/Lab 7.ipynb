```python
# import the needed libraries
import cv2
import numpy as np
import matplotlib.pyplot as plt

# load the image
img = cv2.imread('blocks.png', 0)

# apply the sobel operator
sobelx = cv2.Sobel(img, cv2.CV_64F, 1, 0, 3)
sobely = cv2.Sobel(img, cv2.CV_64F, 0, 1, 3)

# Get absolute value
sobelxAbs = cv2.convertScaleAbs(sobelx)
sobelyAbs = cv2.convertScaleAbs(sobely)

# display the result
plt.imshow(img, 'gray')
plt.show()

plt.imshow(sobelx, 'gray')
plt.show()

plt.imshow(sobely, 'gray')
plt.show()

plt.imshow(sobelxAbs, 'gray')
plt.show()

plt.imshow(sobelyAbs, 'gray')
plt.show()
```

```python
# import image
img3 = cv2.imread('blocks.png', 0)

# apply the Laplacian operator
imgLap = cv2.Laplacian(img3, cv2.CV_64F)
lapAbs = cv2.convertScaleAbs(imgLap)

# display the output
plt.imshow(img3, 'gray')
plt.show()

plt.imshow(imgLap, 'gray')
plt.show()

plt.imshow(lapAbs, 'gray')
plt.show()
```

```python
# use GaussianBlur method
img4 = cv2.GaussianBlur(img3, (3, 3), 0)

# apply Laplacian
imgLapG = cv2.Laplacian(img4, cv2.CV_64F)
lapAbsG = cv2.convertScaleAbs(imgLapG)

# Display
plt.imshow(img3, 'gray')
plt.show()

plt.imshow(imgLap, 'gray')
plt.show()

plt.imshow(imgLapG, 'gray')
plt.show()

plt.imshow(lapAbs, 'gray')
plt.show()

plt.imshow(lapAbsG, 'gray')
plt.show()
```
